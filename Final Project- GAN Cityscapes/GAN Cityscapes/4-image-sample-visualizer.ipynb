## 3. Define function to visualize samples

```python
def show_images(images, figsize=(30,20), columns = 5):
  plt.figure(figsize=figsize)
  for i, file in enumerate(images):
      plt.subplot(len(images) / columns + 1, columns, i + 1)
      img = Image.open(file)
      plt.imshow(img)
      plt.axis("off")
  
  plt.show()
```

## 4. Visualize random sample from the full uncleaned dataset

```python
random_sample = np.random.choice(image_files,20)
```

```python
show_images(random_sample)
```

## 5. Visualize sample from the cleaned dataset

```python
random_sample2 = np.random.choice(cityscape_images,20)
```

```python
show_images(random_sample2, figsize=(30,30))
```
